```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load The Dataset

```python
data = pd.read_csv('/content/Dataset(train).csv')
```

```python
data
```

Bar graph: Count of passengers by class

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Pclass')
plt.title('Count of Passengers by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

Histogram: Age distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Age', bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Cluster plot: Fare vs Age

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='Fare', hue='Pclass')
plt.title('Cluster Plot: Fare vs Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

Box plot: Age distribution by class

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Pclass', y='Age')
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

Scatter plot: Age vs Fare

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='Fare', hue='Survived')
plt.title('Scatter Plot: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

Pair plot: Scatter plot matrix for numeric features

```python
sns.pairplot(data=data, hue='Survived', vars=['Age', 'Fare', 'Pclass', 'SibSp', 'Parch'])
plt.suptitle('Pair Plot of Numeric Features', y=1.02)
plt.show()
```

 Count plot: Count of passengers by gender

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Sex', hue='Survived')
plt.title('Count of Passengers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Violin plot: Fare distribution by class

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='Pclass', y='Fare', hue='Survived', split=True)
plt.title('Fare Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

Heatmap: Correlation matrix of numeric features

```python
plt.figure(figsize=(10, 6))
correlation_matrix = data[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']].corr()
sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

Box plot: Age distribution by class and gender

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='Pclass', y='Age', hue='Sex')
plt.title('Age Distribution by Class and Gender')
plt.xlabel('Class')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.show()
```

Swarm plot: Age distribution by class and gender

```python
plt.figure(figsize=(12, 8))
sns.swarmplot(data=data, x='Pclass', y='Age', hue='Sex', dodge=True)
plt.title('Age Distribution by Class and Gender')
plt.xlabel('Class')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.show()
```

 FacetGrid: Fare distribution by class and embarkation point

```python
g = sns.FacetGrid(data, col='Pclass', row='Embarked', height=4)
g.map(sns.histplot, 'Fare', bins=20)
g.set_titles('Fare Distribution in Class {col_name} - Embarked {row_name}')
plt.show()
```

Catplot: Survived vs Age distribution by class and gender

```python
sns.catplot(data=data, x='Pclass', y='Age', hue='Sex', col='Survived', kind='violin', split=True, height=6)
plt.suptitle('Age Distribution by Class, Gender, and Survival', y=1.02)
plt.show()
```
